```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# 1. Definindo as equaes de Lorenz
# sigma, rho e beta so parmetros que determinam o comportamento do sistema
def lorenz_system(current_state, t, sigma, rho, beta): # Definindo o campo vetorial
    x, y, z = current_state
    dxdt = sigma * (y - x)
    dydt = x * (rho - z) - y
    dzdt = x * y - beta * z
    return [dxdt, dydt, dzdt]

# 2. Condies iniciais e parmetros
params = (10.0, 28.0, 8.0/3.0)  # Valores clssicos para caos
initial_state = [1.0, 1.0, 1.0]
t = np.linspace(0, 50, 10000)   # Tempo de integrao

# 3. Resolvendo a EDO (Integrao Numrica)
solution = odeint(lorenz_system, initial_state, t, args=params)
x, y, z = solution.T
# odeint recebe a funo com as EDOs, onde o sistema comea e os pontos no tempo
# e ento resolve a integrao e devolve uma matriz com as posies x, y e z.

# 4. Plotando o resultado
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, lw=0.5, color='purple') # Pega os resultados da integrao e desenha a trajetria no espao de fases
ax.set_title("Lorenz Attractor - Phase Space")
plt.show()
```
